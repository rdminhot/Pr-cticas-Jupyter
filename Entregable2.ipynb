DIFICULTAD PARA TRAGAR: 2 si tienen problemas para tragar y 1 si no lo tienen.

```python
cancerdf = cancerdf.copy()
```

```python
cancerdf["GENDER"] = cancerdf["GENDER"].replace({"M" : "Male" , "F" : "Female"})
```

```python
for column in cancerdf.columns:
    cancerdf[column] = cancerdf[column].replace({2: "Yes" , 1 : "No"})
```

```python
cancerdf.head(10)
```

## Se Crea Tabla con los Casos Positivos

```python
cancerdf_pos= cancerdf[cancerdf["LUNG_CANCER"] == "YES"]
cancerdf_pos.head()
```

## Paleta para Visualización


```python
sns.set_style("whitegrid")
sns.set_context("poster",font_scale = .7)

palette = ["#1d7874","#679289","#f4c095","#ee2e31","#ffb563","#918450","#f85e00","#a41623","#9a031e","#d6d6d6","#ffee32","#ffd100","#333533","#202020"]

# sns.palplot(sns.color_palette(palette))
# plt.show()
```

## Analizando los Casos Por Genero Y Edad

```python
_, axs = plt.subplots(2,1,figsize=(20,10),sharex=True,sharey=True)
plt.tight_layout(pad=4.0)
sns.histplot(cancerdf_pos[cancerdf_pos["GENDER"]=="Male"]["AGE"],color=palette[1],kde=True,ax=axs[0],bins=20,alpha=1,fill=True)

axs[0].lines[0].set_color(palette[2])
axs[0].set_title("\nCasos Masculinos Positivos Distribucion por Edades\n",fontsize=20)
axs[0].set_xlabel("Age")
axs[0].set_ylabel("Quantity")

sns.histplot(cancerdf_pos[cancerdf_pos["GENDER"]=="Female"]["AGE"],color=palette[12],kde=True,ax=axs[1],bins=20,alpha=1,fill=True)
axs[1].lines[0].set_color(palette[11])
axs[1].set_title("\nCasos Femeninos Positivos Distribucion por Edades\n",fontsize=20)
axs[1].set_xlabel("Age")
axs[1].set_ylabel("Quantity")

sns.despine(left=True, bottom=True)
plt.show()

```

## Analizando La Distribucion de Casos por Genero 

## Distribucion en % de Fumadores Y no Fumadores



```python
plt.subplots(figsize=(9, 9))

labels = "Male","Female"
size = .8



wedges, texts, autotexts = plt.pie([len(cancerdf_pos[cancerdf_pos["GENDER"]=="Male"]["GENDER"]),
                                    len(cancerdf_pos[cancerdf_pos["GENDER"]=="Female"]["GENDER"])],
                                    explode = (0.1,0),
                          textprops=dict(size= 25, color= "white"),
                                    autopct="%.2f%%", 
                                    pctdistance = 0.5,
                                    radius=.9, 
                                    colors = ["#a2c4c9","#e69138"], 
                                    shadow = True,
                                    wedgeprops=dict(width = size, edgecolor = "white", 
                                    linewidth = 5),
                                    startangle = 0)

plt.legend(wedges, labels, title="Category",loc="center left",bbox_to_anchor=(1, 0, 0.5, 1))
plt.title("\nPositive Cases' Gender Distribution",fontsize=20)
plt.show()
plt.subplots(figsize=(9, 9))

labels = "Yes","No"
size = .8



wedges, texts, autotexts = plt.pie([len(cancerdf_pos[cancerdf_pos["SMOKING"]=="Yes"]["SMOKING"]),
                                    len(cancerdf_pos[cancerdf_pos["SMOKING"]=="No"]["SMOKING"])],
                                    explode = (0.1,0),
                                    textprops=dict(size= 25, color= "white"),
                                    autopct="%.2f%%", 
                                    pctdistance = 0.5,
                                    radius=.9, 
                                    colors = ["#a2c4c9","#e69138"], 
                                    shadow = True,
                                    wedgeprops=dict(width = size, edgecolor = "white", 
                                    linewidth = 5),
                                    startangle = 0)

plt.legend(wedges, labels, title="Category",loc="center left",bbox_to_anchor=(1, 0, 0.5, 1))
plt.title("\nPositive Cases' SMOKING Distribution",fontsize=20)
plt.show()          
```
